```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load data
df = pd.read_csv("data/studentdata_featured.csv")

# One-Hot Encoding
df = pd.get_dummies(df, columns=['Gender', 'SchoolType'], drop_first=True)

# Ordinal Encoding
df['ParentalEducation_Encoded'] = df['ParentalEducation'].map({
    '<HS': 1, 'HS': 2, 'SomeCollege': 3, 'Bachelors+': 4
})

# Label Encoding for targets
le1 = LabelEncoder()
df['PerformanceCategory_Encoded'] = le1.fit_transform(df['PerformanceCategory'])

le2 = LabelEncoder()
df['RiskLevel_Encoded'] = le2.fit_transform(df['RiskLevel'])

# Drop original columns
df = df.drop(columns=['ParentalEducation', 'PerformanceCategory', 
                      'RiskLevel'])

# Save
df.to_csv("data/studentdata_encoded.csv", index=False)
print("Saved to: data/studentdata_encoded.csv")
print(f"\nShape: {df.shape}")
print("\nTargets:")
print("  • PerformanceCategory_Encoded (Q1)")
print("  • RiskLevel_Encoded (Q3)")
```
