```python
import pandas as pd
```

```python
# Load cleaned data
df = pd.read_csv("data/studentdata_clean.csv")
```

```python
# Classify students into High / Medium / Low performers based on GPA
def categorize_performance(gpa):
    if gpa >= 3.5:
        return 'High'
    elif gpa >= 2.5:
        return 'Medium'
    else:
        return 'Low'

df['Performance_Category'] = df['GPA'].apply(categorize_performance)
print("Added column: Performance_Category (based on GPA)")
```

```python
#Total and Average Test Score
df['Total_Test_Score'] = df['TestScore_Math'] + df['TestScore_Reading'] + df['TestScore_Science']
df['Average_Test_Score'] = df['Total_Test_Score'] / 3
print("Added columns: Total_Test_Score, Average_Test_Score")
```

```python
#Engagement Score
#A weighted formula considering study habits, attendance, and activities
df['Engagement_Score'] = (df['StudyHours'] * 2) + (df['AttendanceRate'] * 10) + (df['Extracurricular'] * 3)
print("Added column: Engagement_Score (student activity & attendance)")
```

```python
#Support System Score
#Measures the student's social and family support
df['Support_System_Score'] = df['ParentSupport'] + df['Extracurricular'] + (1 - df['PartTimeJob'])
print("Added column: Support_System_Score (family & school support)")
```

```python
# Risk Flag for At-Risk Students
# Flag students with low GPA or poor attendance
df['At_Risk'] = ((df['GPA'] < 2.0) | (df['AttendanceRate'] < 0.85)).astype(int)
print("Added column: At_Risk (1 = potential dropout risk)")
```

```python
output_path = "data/studentdata_featured.csv"
df.to_csv(output_path, index=False)

print(f"\nFeature Engineering dataset saved as: {output_path}")
```
