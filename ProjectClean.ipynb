```python
# Check value ranges for important columns
range_checks = {
    'Age': (14, 18),
    'Grade': (9, 12), 
    'GPA': (0.0, 4.0),
    'AttendanceRate': (0.70, 1.00),
    'StudyHours': (0, 4),
    'FreeTime': (1, 5),
    'GoOut': (1, 5),
    'TestScore_Math': (0, 100),
    'TestScore_Reading': (0, 100),
    'TestScore_Science': (0, 100)
}

issues_found = False
for col, (min_val, max_val) in range_checks.items():
    if col in df.columns:
        invalid_count = df[(df[col] < min_val) | (df[col] > max_val)].shape[0]
        if invalid_count > 0:
            print(f"{col}: {invalid_count} values outside valid range ({min_val}–{max_val})")
            issues_found = True  
            
if not issues_found:
    print("All column values are within valid ranges.")
```

```python
# Remove columns that won't help ML performance
columns_to_remove = ['Grade', 'Romantic', 'Locale', 'Race', 'InternetAccess', 'GoOut', 'FreeTime',]

for col in columns_to_remove:
    if col in df.columns:
        df.drop(col, axis=1, inplace=True)
        print(f"Removed: {col}")
```
